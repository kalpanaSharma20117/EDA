```python
# M aagya ab isko replace krne k lie koi function bnaate h>>
# 1 MB = 1024 kb

def size_process(item):
    if str(item)[-1] == "M":
        res = float(str(item).replace("M", ""))
        res = res * 1024
        return res
    elif str(item)[-1] == "k":
        res = float(str(item).replace("k", ""))
        return res
    else:
        return str(np.nan)
        
```

```python
size_process("19M")
```

```python
df_copy['Size'] = df_copy['Size'].apply(size_process)
```

```python
#internal assignment >
#prepare android version, replace w, keep values with 1 decimal point, replcae dash>>
def clean_android(item):
    if pd.isnull(item) or item.strip() == "":
        return np.nan
    
    if "-" in item:
        item = item.split("-")[-1]
    item = item.strip()#Removes any extra spaces again, especially useful after the split.

    parts = item.split(".")
    if len(parts) >=2 :
        return f"{parts[0]}.{parts[1]}"
    else:
        return item
    
df_copy['Android Ver'] = df_copy['Android Ver'].apply(clean_android)
```
